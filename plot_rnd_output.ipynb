```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from scipy.special import softmax

# standardized_int_rwd = [
#     [ 0.69527124,  0.25369398, 15.9830697 , -1.61889592],
#     [ 0.94144165, -2.76592344, 18.42739069,  0.84786037],
#     [-6.20783837, -2.08363863, -2.64843787, -1.39636001],
#     [ 3.78511988, -1.96097384,  4.69459714, -2.72659269],]

standardized_int_rwd = np.genfromtxt("log/rnd_log.csv", delimiter=",")
standardized_int_rwd = standardized_int_rwd[:,1:]

k = -0.5
reverse_norm_std_int_rwd = softmax(k*np.array(standardized_int_rwd), axis=1)
print(np.sum(reverse_norm_std_int_rwd, axis=1))
for i in range(len(reverse_norm_std_int_rwd)):
    for w in reverse_norm_std_int_rwd[i]:
        print(f"{w:.2f}", end=" ")
    print()
plt.suptitle("Weight-like RND output")
for i, line in enumerate(reverse_norm_std_int_rwd):
    plt.subplot(f"22{i+1}")
    plt.title(f"base{i}")
    plt.xlabel("model")
    plt.ylim(0, 1)
    plt.plot(line, "--bs")
plt.show()
```
