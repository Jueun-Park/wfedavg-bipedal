```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from scipy.special import softmax
standardized_int_rwd = [
    [ 6.34856915, -2.16217204, -1.96735414, -2.9913503 ],
    [ 6.9297269,  -2.23938597, -2.92667986, -1.97417563],
    [ 9.18912909,  4.05392035, -4.00278517, -1.20934996],
    [ 9.67959552,  3.76962278, -3.74632527, -2.17500249]]
reverse_norm_std_int_rwd = softmax(-np.array(standardized_int_rwd))

plt.suptitle("f(int_rwd)")
for i, line in enumerate(reverse_norm_std_int_rwd):
    plt.subplot(f"22{i+1}")
    plt.title(f"base{i}")
    plt.xlabel("model")
    plt.plot(line, "--gs")
plt.show()
```
